```python
import numpy as np
```

```python
Q1= data['NumDots'].quantile(0.25)
Q3= data['NumDots'].quantile(0.75)
IQR= Q3 - Q1
```

```python
print(IQR)
```

```python
lower_bound = Q1 - 1.5 * IQR
print(lower_bound)
upper_bound = Q3 + 1.5 * IQR
print(upper_bound)
```

```python
# find the outliers
outliers = data[(data['NumDots'] < lower_bound) | (data['NumDots'] > upper_bound)]
print(outliers)
```

```python
outliers.head(15)
```

```python
from scipy.stats import zscore
```

```python
data["Z_scores"] = zscore(data["NumDots"])

# Identify outliers (Z-score threshold: ±3)
z_scoreOutliers = data[(data["Z_scores"] < -3) | (data["Z_scores"] > 3)]
print(z_scoreOutliers)
```
